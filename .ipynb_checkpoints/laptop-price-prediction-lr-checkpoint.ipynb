# **Data Cleaning**

first we drop laptop_id and product

```python
df.drop(['laptop_ID','Product'],axis = 1, inplace = True)
```

```python
df['Company'].value_counts()
```

```python
df['Inches'].value_counts()
```

<h2>ScreenResolution</h2>

Next ScreenResolution, We can separate it into several columns to make it useful
1. Resolution
2. TouchScreen or no?
3. Type IPS Panel or IPS Panel Retina Display

```python
df['ScreenResolution'].value_counts()
```

```python
df['Resolution'] = df['ScreenResolution'].str.extract(r'(\d+x\d+)') 
df['Resolution'].value_counts()
```

```python
df['ScreenType'] = df['ScreenResolution']
```

```python
def simplify_screentype(entry):
    pattern = r'IPS Panel Retina|IPS Panel'
    match = re.search(pattern, entry)
    if match:
        return match.group()
    else :
        return None
```

```python
df['ScreenType'] = df['ScreenType'].apply(simplify_screentype)
df['ScreenType'].value_counts()
```

```python
df['TouchScreen'] = df['ScreenResolution']
```

```python
def simplify_touchscreen(entry):
    pattern = r'TouchScreen'
    match = re.search(pattern, entry,re.IGNORECASE)
    if match:
        return 'Yes'
    else :
        return 'No'
```

```python
df['TouchScreen'] = df['TouchScreen'].apply(simplify_touchscreen)
df['TouchScreen'].value_counts()
```

```python
df.drop('ScreenResolution', axis = 1, inplace = True)
df.head(2)
```

<h2>CPU</h2>

We will divide the CPU data into two columns:
1. CPU
2. Frequency

```python
df['Cpu_Frequency'] = df['Cpu'].str.extract(r'(\d+(?:\.\d+)?GHz)')
df['Cpu_Frequency'] = df['Cpu_Frequency'].str.replace('GHz', '')
df.rename(columns={'Cpu_Frequency': 'Cpu_Frequency(GHz)'}, inplace=True)
df['Cpu_Frequency(GHz)'].value_counts()
```

```python
df['Cpu']= df['Cpu'].str.replace(r'(\d+(?:\.\d+)?GHz)', '', regex=True)
```

```python
def simplify_cpu(entry):
    pattern = r'Intel Core i[357]|AMD|Samsung'
    match = re.search(pattern, entry,re.IGNORECASE)
    if match:
        return match.group()
    else :
        return 'Other Intel'
```

```python
df['Cpu'] = df['Cpu'].apply(simplify_cpu)
df['Cpu'].value_counts()
```

<h2>RAM</h2>
We will remove Ram from each row

```python
df['Ram'] = df['Ram'].str.replace('GB', '')
df.rename(columns={'Ram': 'Ram(GB)'}, inplace=True)
df['Ram(GB)'].value_counts()
```

<h2>Memory</h2>

If we look, there is a large amount of data that can be divided :
1. Type memory
2. Memory capacity

```python
df['Memory'].value_counts()
```

```python
def convert_memory_to_description(memory_type):
    if "SSD" in memory_type and "HDD" in memory_type:
        return "SSD and HDD"
    elif "Flash Storage" in memory_type and "HDD" in memory_type:
        return "Flash Storage and HDD"
    elif "SSD" in memory_type and "Hybrid" in memory_type:
        return "SSD and Hybrid"
    elif "Flash Storage" in memory_type:
        return "Flash Storage"
    elif "HDD" in memory_type:
        return "HDD"
    elif "SSD" in memory_type:
        return "SSD"
    elif "Hybrid" in memory_type:
        return "Hybrid"
    else:
        return memory_type
```

```python
df['TypeMemory'] = df['Memory'].apply(convert_memory_to_description)
df['TypeMemory'].value_counts()
```

```python
df['Memory'] = df['Memory'].str.replace('GB', '')
df['Memory'] = df['Memory'].str.replace('1TB', '1000')
df['Memory'] = df['Memory'].str.replace('1.0TB', '1000')
df['Memory'] = df['Memory'].str.replace('2TB', '2000')
```

```python
def sum_memory(memory_str):
    sizes = re.findall(r'\d+', memory_str)
    sizes = [int(size) for size in sizes]
    total_gb = sum(sizes)
    
    return total_gb
```

```python
df['Memory']= df['Memory'].apply(sum_memory)
```

```python
df.rename(columns={'Memory': 'Memory(GB)'}, inplace=True)
df['Memory(GB)'].value_counts()
```

<h2>GPU</h2>
For GPU, we will categorize them into Intel, Nvidia, AMD, and ARM

```python
def simplify_gpu(entry):
    pattern = r'Intel|AMD|Nvidia|ARM'
    match = re.search(pattern, entry,re.IGNORECASE)
    if match:
        return match.group()
    else :
        return entry
```

```python
df['Gpu'] = df['Gpu'].apply(simplify_gpu)
df['Gpu'].value_counts()
```

<h2>Weight</h2>
We will remove "kg" from each line, then categorize them according to the standard as follows: light, medium, heavy.

```python
df['Weight'] = df['Weight'].str.replace('kg', '')
df['Weight'] = pd.to_numeric(df['Weight'])
df['Weight'] = pd.qcut(df['Weight'], q=3, labels=['Light', 'Medium', 'Heavy'])
df['Weight'].value_counts()
```

# Models & Evaluation

```python
df.dtypes
```

```python
cat_col = ['Company', 'TypeName', 'Cpu', 'Gpu','OpSys', 'Weight', 'Resolution', 'ScreenType',
       'TouchScreen', 'Cpu_Frequency(GHz)', 'TypeMemory']
cat_col
```

```python
pd.get_dummies(df, columns=cat_col, drop_first=True).shape
```

```python
label_encoder= LabelEncoder()
for col in cat_col :
    df[col] = label_encoder.fit_transform(df[col].astype(str))
```

```python
df.head()
```

```python
x=df.drop('Price_euros', axis = 1).values
y=df['Price_euros'].values
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
train_score = lm.score(x_train, y_train)
test_score = lm.score(x_test, y_test)
print(f'Training R-squared: {train_score:.2f}')
print(f'Test R-squared: {test_score:.2f}')
```

```python
y_pred = lm.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Harga Aktual")
plt.ylabel("Harga Prediksi")
plt.title("Perbandingan Harga Aktual dan Harga Prediksi (Data Pengujian)")
plt.show()
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error (MSE):', mse)
print('R-squared (R^2):', r2)
```

```python
res = y_test - y_pred
```

```python
plt.scatter(y_pred, res)
plt.axhline(y=0, color='red', linestyle='--')  # Garis referensi di nol (sebagai acuan jika residuals = 0)
plt.xlabel('Predicted Prices')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
feature = ['Company', 'TypeName', 'Inches', 'Cpu', 'Ram(GB)', 'Memory(GB)', 'Gpu',
       'OpSys', 'Weight', 'Resolution', 'ScreenType',
       'TouchScreen', 'Cpu_Frequency(GHz)', 'TypeMemory']
cat_col
coefficients = lm.coef_
# Buat DataFrame untuk menyimpan feature importance
feature_importance_df = pd.DataFrame({'cat_col': feature, 'Coefficient': coefficients})

# Urutkan feature importance dari nilai absolut tertinggi ke terendah
feature_importance_df['Abs_Coefficient'] = np.abs(feature_importance_df['Coefficient'])
feature_importance_df = feature_importance_df.sort_values(by='Abs_Coefficient', ascending=False)

# Tampilkan feature importance dalam bentuk tabel
print(feature_importance_df)
```

```python
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['cat_col'], feature_importance_df['Coefficient'])
plt.xlabel('Coefficient')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.show()
```
